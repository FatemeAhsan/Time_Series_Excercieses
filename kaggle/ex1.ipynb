```python
df = pd.read_csv('train.csv')
df
```

```python
df = df.set_index('date')
df
```

```python
df = df.set_index(['store_nbr', 'family'], append=True)
df
```

```python
df = df.groupby('date').mean()['sales']
df
```

```python
df = df.to_frame()
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df['time'] = np.arange(len(df.index)) 

X = df.loc[:, ['time']]
y = df.loc[:, 'sales']

model = LinearRegression()
model.fit(X, y)

y_pred = pd.Series(model.predict(X), index=X.index)
```

```python
plt.figure(figsize=[10, 3])
ax = y.plot(alpha=0.5, color=np.random.rand(3))
ax = y_pred.plot(ax=ax, linewidth=3, color=np.random.rand(3))
```

```python
df['lag_1'] = df['sales'].shift(1)

X = df.loc[:, ['lag_1']].dropna()
y = df.loc[:, 'sales']
y, X = y.align(X, join='inner')

model = LinearRegression()
model.fit(X, y)

y_pred = pd.Series(model.predict(X), index=X.index)
```

```python
fig, ax = plt.subplots()
ax.plot(X['lag_1'], y, '.', color=np.random.rand(3))
ax.plot(X['lag_1'], y_pred, color=np.random.rand(3))
ax.set(aspect='equal', ylabel='sales', xlabel='lag_1', title='Lag Plot of Average Sales')
```
